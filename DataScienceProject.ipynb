**1. Feature Selection**

- Tạo ra cột dữ liệu Goal diff và tính hệ số tương quan với nó
- Dựa trên hệ số tương quan của các feature với dữ liệu Goal Differance, ta thực hiện feature selection bằng cách chọn 10 giá trị có hệ số tương quan cao nhất với Goal Differance.

```python
from sklearn import metrics
from sklearn import preprocessing

def CreateTargetVariable(data):
  # Tính ma trận độ tương quan
  data['Goal Difference'] = data['Score Home'] - data['Score Away']
  corr_matrix = data.drop(['Team Home','Team Away','Score Home','Score Away'], axis=1).corr()


def CaculateHSTQ(data):
  corr_matrix = data.drop(['Team Home','Team Away','Score Home','Score Away'], axis=1).corr()
  # Lấy ra 12 trường có độ tương quan cao nhất với hiệu số bàn thắng
  top_10_corr = corr_matrix['Goal Difference'].abs().nlargest(12)
  print(top_10_corr)
  print("\n")
  #loại bỏ bản thân cột Goal Diff và cột Match Result ra khỏi feature được chọn, còn lại chọn 10 cột
  columns_to_keep = top_10_corr.index.drop(top_10_corr.index[0:2]).tolist()
  return data[columns_to_keep], data['Goal Difference']

def Create2ndTargetVariable(data):
  # Create new feature tên 'Match Result'
  #le = preprocessing.LabelEncoder()
  data['MatchResult'] = data['Goal Difference'].apply(lambda x: 2 if x > 0 else -2 if x < 0 else 0)
  #data['MatchResult'] = le.fit_transform(data['MatchResult'])

#Tạo thêm cột dữ liệu Goal Diff trên 2 dataset
CreateTargetVariable(big_data)
CreateTargetVariable(small_data)
print("\n")
#Tạo thêm cột dữ liệu Match Result trên 2 dataset
Create2ndTargetVariable(big_data)
Create2ndTargetVariable(small_data)
print("\n")
#lấy ra dữ liệu là data variable từ feature được chọn và data của target variable
X1_big = CaculateHSTQ(big_data)
X1_small = CaculateHSTQ(small_data)
```

**1. Linear Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold

def ModelAccuracy(data, y, model):
    test_size = 0.2
    random_state = 30

    # Chia dữ liệu thành tập train và tập test
    X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=test_size, random_state=random_state)

    # Huấn luyện mô hình Linear Regression với tập train
    model.fit(X_train, y_train)

    # Dự đoán trên tập test
    y_pred = model.predict(X_test)

    # Tính toán các metric
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    # In ra kết quả
    print("R-squared score: ", r2)
    print("Mean Squared Error: ", mse)
    print("Root Mean Squared Error: ", rmse)

    return model, r2, mse, rmse

LRmodel = LinearRegression()
LR_Big = ModelAccuracy(X1_big[0],X1_big[1],LRmodel)
print("\n")
LR_Small = ModelAccuracy(X1_small[0],X1_small[1],LRmodel)
```

```python
def plot_performance(big_scores, small_scores):
    metrics = ['R-squared', 'Mean Squared Error', 'Root Mean Squared Error']

    fig, ax = plt.subplots(1, 3, figsize=(16, 6))

    for i, metric in enumerate(metrics):
        scores = [big_scores[i], small_scores[i]]
        ax[i].bar(['Big Dataset', 'Small Dataset'], scores, color=['blue', 'orange'])
        ax[i].set_title(metric)
        ax[i].set_ylabel('Score')
        #ax[i].set_ylim(min(scores) - 0.1 * np.abs(min(scores)), max(scores) + 0.1 * np.abs(max(scores)))

    plt.tight_layout()
    plt.show()
```

1. Winsorizing method

```python
def HandleOutliner(data):
  numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()
  for col in numeric_columns:
      q1 = data[col].quantile(0.25)
      q3 = data[col].quantile(0.75)
      iqr = q3 - q1
      lower_bound = q1 - (1.5 * iqr)
      upper_bound = q3 + (1.5 * iqr)
      data[col] = np.where(data[col] < lower_bound, data[col].mean(), data[col])
      data[col] = np.where(data[col] > upper_bound, data[col].mean(), data[col])

  # # Kiểm tra lại sau khi đã xử lý giá trị ngoại lai
  # outliers = []
  # for col in numeric_columns:
  #     q1 = data[col].quantile(0.15)
  #     q3 = data[col].quantile(0.85)
  #     iqr = q3 - q1
  #     lower_bound = q1 - (1.5 * iqr)
  #     upper_bound = q3 + (1.5 * iqr)
  #     outliers.extend(data[(data[col] < lower_bound) | (data[col] > upper_bound)].index.tolist())


bigdata_copy = big_data.copy()
smalldata_copy = small_data.copy()

HandleOutliner(bigdata_copy)
X1copy_big = CaculateHSTQ(bigdata_copy)
print("\n")
HandleOutliner(smalldata_copy)
X1copy_small = CaculateHSTQ(smalldata_copy)
```

2. Gaussian Distribution

```python
##### Assuming Age follows A Gaussian Distribution we will calculate the boundaries which differentiates the outliers
from scipy.stats import norm

def replace_outliers_with_boundaries(data):
    data_column = data.select_dtypes(include=[np.number]).columns.tolist()
    for column in data_column:
        upper_boundary = data[column].mean() + 3 * data[column].std()
        lower_boundary = data[column].mean() - 3 * data[column].std()

        # Replace outliers with boundaries
        data[column] = np.where(data[column] > upper_boundary, upper_boundary, data[column])
        data[column] = np.where(data[column] < lower_boundary, lower_boundary, data[column])


bigdata_copy2 = big_data.copy()

smalldata_copy2 = small_data.copy()

replace_outliers_with_boundaries(bigdata_copy2)
X1copy_big2 = CaculateHSTQ(bigdata_copy2)
print("\n")
replace_outliers_with_boundaries(smalldata_copy2)
X1copy_small2 = CaculateHSTQ(smalldata_copy2)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

import numpy as np

def normalize_min_max(X):
    min_vals = np.min(X, axis=0)
    max_vals = np.max(X, axis=0)
    normalized_X = (X - min_vals) / (max_vals - min_vals)
    return normalized_X, min_vals, max_vals


# Áp dụng chuẩn hóa cho tập dữ liệu X_big_train
X_big_train_normalized, min_vals, max_vals = normalize_min_max(X1_big[0])


def normalize_min_max(y):
    min_val = np.min(y)
    max_val = np.max(y)
    normalized_y = (y - min_val) / (max_val - min_val)
    return normalized_y, min_val, max_val


# # Chuẩn hóa tập y
# y_big_train_normalized, y_min, y_max = normalize_min_max(y_big)

# Áp dụng chuẩn hóa cho tập y_big_train
y_big_train_normalized, y_min, y_max = normalize_min_max(X1_big[1])


def gradient_descent(X, y, alpha, num_iterations):
    num_samples, num_features = X.shape
    theta = np.zeros(num_features)  # Khởi tạo tham số theta ban đầu

    for i in range(num_iterations):
        y_pred = np.dot(X, theta)  # Dự đoán giá trị y

        # Tính toán gradient của hàm mất mát (MSE)
        gradient = np.dot(X.T, (y_pred - y)) / num_samples

        # Cập nhật tham số theta
        theta = theta - alpha * gradient

    return theta


def evaluate_model(X, y, alpha_values, num_iterations):
    r2_scores = []
    mse_scores = []
    mae_scores = []
    best_r2_score = -np.inf
    best_mse_score = np.inf
    best_mae_score = np.inf
    best_alpha_r2 = None
    best_alpha_mse = None
    best_alpha_mae = None

    for alpha in alpha_values:
        # Tối ưu hóa mô hình bằng Gradient Descent
        theta = gradient_descent(X, y, alpha, num_iterations)

        # Dự đoán giá trị y
        y_pred = np.dot(X, theta)

        # Đánh giá mô hình
        r2 = r2_score(y, y_pred)
        mse = mean_squared_error(y, y_pred)
        mae = mean_absolute_error(y, y_pred)

        r2_scores.append(r2)
        mse_scores.append(mse)
        mae_scores.append(mae)

        # Kiểm tra và cập nhật điểm số tốt nhất và alpha tương ứng
        if r2 > best_r2_score:
            best_r2_score = r2
            best_alpha_r2 = alpha
        if mse < best_mse_score:
            best_mse_score = mse
            best_alpha_mse = alpha
        if mae < best_mae_score:
            best_mae_score = mae
            best_alpha_mae = alpha

    # Hiển thị và in ra điểm số tốt nhất và alpha tương ứng
    print("Best R2 Score:", best_r2_score)
    print("Best Alpha (R2):", best_alpha_r2)
    print("Best MSE Score:", best_mse_score)
    print("Best Alpha (MSE):", best_alpha_mse)
    print("Best MAE Score:", best_mae_score)
    print("Best Alpha (MAE):", best_alpha_mae)

    # Trực quan hóa ảnh hưởng của alpha đến R2, MSE và MAE
    plt.plot(alpha_values, r2_scores, label='R2')
    plt.plot(alpha_values, mse_scores, label='MSE')
    plt.plot(alpha_values, mae_scores, label='MAE')
    plt.xlabel('Alpha')
    plt.ylabel('Score')
    plt.title('Effect of Alpha on Model Performance')
    plt.legend()
    plt.show()


# Tham số alpha và số lần lặp
num_iterations = 1000
alpha_values = [0.01, 0.1, 0.5, 1.0] # trên 1 là bị lỗi
# Áp dụng vào mô hình và đánh giá
evaluate_model(X_big_train_normalized,y_big_train_normalized, alpha_values, num_iterations)
```

Để tiếp tục, ta tạm chấp nhận giả định sau:

**Sử dụng hàm sigmoid để dự toán tỉ lệ thắng/hòa/thua dự trên hiệu số bàn thắng kỳ vọng**

- Sử dụng hàm sigmoid tính tỉ lệ thắng/thua/hòa của 1 trận đấu:

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-(x)))

def normalize(x, new_min, new_max):
    min_value = 0
    max_value = 0.5
    normalized_x = (x - min_value) * (new_max - new_min) / (max_value - min_value) + new_min
    return normalized_x

def predict_outcome_probabilities(goal_difference_expected):
    win_prob = sigmoid(goal_difference_expected)
    lose_prob = sigmoid(-goal_difference_expected)
    draw_prob = sigmoid(-abs(goal_difference_expected))
    draw_prob = draw_prob / 5 * 4.2
    #chuẩn hóa dải draw_prob từ [0,0.5] thành [0, 0.42] theo phương pháp MinMaxscaler. Lý do cho con số 0.42 là vì 0.42 / (1 + 0.42) xấp xỉ 30%, tức giá trị MaxDrawProbability đã xác định trước đó.
    total = win_prob + lose_prob + draw_prob
    return win_prob/total , draw_prob/total , lose_prob/total
```

```python
def predict_outcome_probabilities_array(goal_difference_expected):

    win_prob = sigmoid(goal_difference_expected)
    lose_prob = sigmoid(-goal_difference_expected)
    draw_prob = sigmoid(-np.abs(goal_difference_expected))
    draw_prob = draw_prob / 5 * 4.2

    total = win_prob + lose_prob + draw_prob

    win_prob_normalized = win_prob / total
    draw_prob_normalized = draw_prob / total
    lose_prob_normalized = lose_prob / total

    return win_prob_normalized, draw_prob_normalized, lose_prob_normalized

# Tạo dữ liệu data
data_GDE = np.linspace(G_min,G_max,1000)

# Tính toán giá trị a, b, c
a, b, c = predict_outcome_probabilities_array(data_GDE)

# Vẽ biểu đồ đường
plt.plot(data_GDE, a, label='Win probability')
plt.plot(data_GDE, b, label='Draw probability')
plt.plot(data_GDE, c, label='Lose probability')

# Đặt tên cho các trục và tiêu đề biểu đồ
plt.xlabel('Goal Difference Expectation')
plt.ylabel('Probability Value')
plt.title('Plot of Win/Draw/Lose probability based on Goal Difference Expectation')
plt.axvline(x=0, color='black', linestyle='--')
# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```
